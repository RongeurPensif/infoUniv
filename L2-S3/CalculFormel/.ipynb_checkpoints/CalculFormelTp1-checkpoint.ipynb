```python
somme = 0
for i in xrange ( 10 ) :
    aireRectangle = (pi/10) * sin (i*pi/10)
    somme += aireRectangle
    print('Aire rectangle n°{} = {}'.format(i+1,numerical_approx(aireRectangle)))
print ( '\nValeur approchée de I = {}'.format( numerical_approx ( somme )) )
```

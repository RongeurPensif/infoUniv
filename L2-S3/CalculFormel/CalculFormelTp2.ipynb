```python
#Exo 4
def Fibrec ( n ) :
    if n < 2 :
        return 1
    else :
        return Fibrec (n-1) + Fibrec (n-2)

def Fibit ( n ) :
    liste = []
    for i in range ( n + 1 ) :
        if i < 2 :
            liste.append(1)
        else :
            liste.append(liste[i-1]+liste[i-2])
    return liste[n]



print('Methode Itérative : ')
timeit('Fibit(1)')

print('\nMethode Récursive : ' )
timeit ('Fibrec(1)')

# 2 µs = 2 *10^-6 s < 0.9 s
```

```python
#Exo 6
def QtChaleur ( n ) :
    c0 = 10
    cn = c0
    r = 1.5
    if n == 0 :
        return c0
    for i in range ( n ) :
        cn = cn + r
    return cn

for i in range (12) :
    print("U{} = {}".format(i, QtChaleur (i)))

#Finir pour la semaineprochaine
```
